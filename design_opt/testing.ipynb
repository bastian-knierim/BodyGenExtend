```python
import numpy as np
import quaternion

def add_rotation(q_curr):
    axes = ['x','y','z']
    axis = np.random.choice(axes)
    q = np.quaternion(*q_curr[[3, 0, 1, 2]])
    
    half = np.pi / 4
    if axis == 'x':
        q_rot = np.quaternion(np.cos(half), np.sin(half), 0, 0)
    elif axis == 'y':
        q_rot = np.quaternion(np.cos(half), 0, np.sin(half), 0)
    elif axis == 'z':
        q_rot = np.quaternion(np.cos(half), 0, 0, np.sin(half))

    q_new = q_rot * q
    return np.array([q_new.x, q_new.y, q_new.z, q_new.w])


q0 = np.array([0, 0, 0, 1])

# 4 Schritte um z-Achse drehen
q = q0
for i in range(4):
    q = add_rotation(q)
    print(f"Step {i+1}: {q}")
```
